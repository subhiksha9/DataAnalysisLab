Calculate  Petal width mean ,median ,SD,Variance  for the species setosa

```python
s=iris[iris['species']=='Iris-setosa']
p=s['petal_width']
print("Mean: ",p.mean())
print("Median: ",p.median())
print("SD: ",p.std())
print("Variance: ",p.var())
```
