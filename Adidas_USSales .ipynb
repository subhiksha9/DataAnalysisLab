```python
import pandas as pd
import numpy as np
df=pd.read_excel('adidas1.xlsx')
df
```

1.	List all the retailers with retailer id

```python
df1=df[['Retailer','Retailer ID']].drop_duplicates()
print(df1)
print(df1.info())
```

2.	List all the retailers in every region.

```python
df2=df.groupby('Region')['Retailer'].unique()
print(df2)
```

3.	List the retailers in every city of a state.

```python
df3=df.groupby(['State','City'])['Retailer'].unique()
print(df3)
```

4.	List the products sold by the retailer in every city

```python
df4=df.groupby(['City','Retailer'])['Product'].unique()
df4
```

5.	Find the total sales of every retailer.

```python
df5=df.groupby('Retailer')['Total Sales'].sum()
df5
```

6.	Find the total sales of the retailers in every city along with profit.

```python
df6=df.groupby(['Retailer','City'])[['Total Sales','Operating Profit']].sum()
df6
```

7.	Find the total sales &profit of each product sold by the retailer.

```python
df7=df.groupby(['Retailer','Product'])[['Total Sales','Operating Profit']].sum()
df7
```
